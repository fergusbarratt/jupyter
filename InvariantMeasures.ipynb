```python
%matplotlib inline
import numpy as np
import scipy.integrate as sp
import matplotlib.pyplot as plt
```

```python
%%capture
def f(x, mu=0.9):
    return 4*mu*x*(1-x)
# t = np.linspace(1, 2, 100)
# sol = sp.odeint(f, 200, t)
# plt.plot(t, sol)
def iterate(func, n_iters, time_points = np.linspace(0.1, 0.105, 20)):
    return_list = []
    for time in time_points:
        initial_val = time
        vals = []
        for _ in range(n_iters):
            initial_val = func(initial_val)
            vals.append(initial_val)
        return_list.append(vals)
    return return_list
sol = iterate(f, 50)
```

```python
plt.plot(np.asarray(sol).T);
```
