```python
%matplotlib inline
def f(x, mu):
    return 4*mu*x*(1-x)

def iterate(func, n_iters, in_val, params):
    res = in_val
    for _ in range(n_iters):
        res = func(res, *params)
    return res

def feigenbaum(func, in_val, epsilon):
    in_iter = iterate(func, 5012, in_val, (0.9,))
    return in_iter

iterate(f, 3, 0.5, (0.9,))
feigenbaum(f, 0.2, 5)
```
