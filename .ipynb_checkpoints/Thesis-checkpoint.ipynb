### Semiclassical

```python
def drive(alphasq, detuning, kappa, g):
    return abs(-1j*np.sqrt(alphasq)*(kappa - 1j*(detuning + (detuning/abs(detuning))*g**2/np.sqrt(detuning**2 + 4 * g**2 * alphasq))))
```

```python
c_q_det = (8.1831-10.5665)
kappa = 0.0024
gamma = 0
coupling_strength = 0.3347
cavity_freq = 10.5665

xiC1= (abs(c_q_det)*kappa)**(3/2)/(3**(3/4)*coupling_strength**2)
chi0=coupling_strength**2/abs(c_q_det)

def drive(A, omega_cavity, omega_drive, sigmaz, det, g, kappa):
  '''self-consistently determine xi from A'''
  def chi(A):
    return sigmaz*(g**2)/np.sqrt(2*g**2*(A**2+sigmaz)+det**2)
  return np.sqrt(A**2 * (1/omega_cavity**2)*(
            (omega_drive**2-(omega_cavity-chi(A))**2)**2 + kappa**2*omega_drive**2))
```

```python
As = np.linspace(0, 100)
Dets = np.linspace(-0.01, 0.05, 100)
Z = np.array([[drive(A, 
                  cavity_freq, 
                  cavity_freq+d, 
                  -1, 
                  c_q_det, 
                  coupling_strength, 
                  kappa) for d in Dets] for A in As])
plt.contour(Dets, As, Z, 100, linewidths=1.0)
plt.colorbar(label='Drive Strength')
plt.title('Semiclassical Dispersive Bistability', loc='right', fontdict={'fontsize': 16, 'verticalalignment': 'bottom'})
plt.xlabel('$\omega_c - \omega_d$')
plt.ylabel('$A^2$')
sb.despine()
```

### Quantum

```python
c_q_det = (8.1831-10.5665) 
kappa = 0.0024 
gamma = 0
coupling_strength = 0.3347  
matrix_size = 10
cavity_freq = 10.5665 

drives = 6.3 
detunings = np.append(np.append(np.linspace(-0.5, 0.6, 10), np.linspace(0.61, 1, 40)), np.linspace(1.1, 1.5, 10))

carmichael_parameters = qo.JaynesCummingsParameters(10, matrix_size).det_params(
                              drive_strengths=6,
                              drive_cavity_detunings=0,
                              qubit_cavity_detunings=0,
                              c_op_params=[1, 1],
                              omega_cavity=10)

bishop_parameters = qo.JaynesCummingsParameters(coupling_strength, matrix_size).det_params(
                              drive_strengths=drives*xiC1,
                              drive_cavity_detunings=detunings*chi0,
                              qubit_cavity_detunings=c_q_det,
                              c_op_params=[kappa],
                              omega_cavity=cavity_freq)

bishop_system = qo.SteadyStateJaynesCummingsModel(
    *bishop_parameters)
carmichael_system = qo.SteadyStateJaynesCummingsModel(
    *carmichael_parameters)

def abs_cavity_field_Q(sys, xvec=np.linspace(-5, 5, 50), yvec=np.linspace(-5, 5, 50)):
    Qs = sys.qps(xvec, yvec)
    return np.array([abs(np.sum(xvec.T.dot(Q))+1j*np.sum(Q.dot(yvec))) for Q in Qs])/34
```

### Analytic Wigner Function
http://iopscience.iop.org/article/10.1088/1464-4266/1/2/005/pdf

```python
eps = 12
x = np.linspace(-8, 8, 100)
y = np.linspace(-8, 8, 100)

def W(alpha, lamb):
    return np.exp(
        -2*np.absolute(alpha)**2) * (
        np.absolute(jv(lamb-1, np.sqrt(-8 * eps * np.conj(alpha)))/
        np.conj(alpha)**((lamb-1)/2)))**2 

w = np.array([[W(x+y*1j, 5) for x in x] for y in y])

plt.contourf(x, y, np.real(w))
plt.colorbar()
plt.title('Wigner Function', loc='right', fontdict={'fontsize': 16, 'verticalalignment': 'bottom'})
plt.xlabel('Re(Q)')
plt.ylabel('Im(Q)')
```

### Spontaneous dressed state polarisation

```python
def alpha_ss(epsilon, kappa=0.5, g=5):
    return (epsilon/kappa)* np.sqrt(1-(g/(2*epsilon))**2)

x = np.linspace(0.1, 4.5, 100)
y = np.vectorize(alpha_ss)(x)

plt.plot(x, y)
```

### Critical Slowing

```python
t_params = np.linspace(0, 10, 200), None
drive_1 = 20
det_1 = 0
drive_2 = 5
det_2 = 0
carmichael_parameters_1 = qo.JaynesCummingsParameters(10, matrix_size).det_params(
                              drive_strengths=drive_1,
                              drive_cavity_detunings=det_1,
                              qubit_cavity_detunings=0,
                              c_op_params=[1, 1],
                              omega_cavity=10)
carmichael_parameters_2 = qo.JaynesCummingsParameters(10, matrix_size).det_params(
                              drive_strengths=drive_2,
                              drive_cavity_detunings=det_2,
                              qubit_cavity_detunings=0,
                              c_op_params=[1, 1],
                              omega_cavity=10)
td_carmichael_system_1 = qo.TimeDependentJaynesCummingsModel(*carmichael_parameters_1, *t_params)
td_carmichael_system_2 = qo.TimeDependentJaynesCummingsModel(*carmichael_parameters_2, *t_params)
```

```python
N = td_carmichael_system_1.a.dag()*td_carmichael_system_1.a
y_1 = td_carmichael_system_1.solve([N])
y_2 = td_carmichael_system_2.solve([N])
plt.plot(t_params[0], y_1.expect[0])
plt.plot(t_params[0], y_2.expect[0])
plt.legend(['drive = {}, det = {}'.format(drive_1, det_1), 'drive = {}, det = {}'.format(drive_2, det_2)])
plt.xlabel('time')
plt.ylabel('$A^2$')
plt.title('Critical Slowing')
```
