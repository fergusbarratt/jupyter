### Semiclassical

```python
def drive(alphasq, detuning, kappa, g):
    return abs(-1j*np.sqrt(alphasq)*(kappa - 1j*(detuning + (detuning/abs(detuning))*g**2/np.sqrt(detuning**2 + 4 * g**2 * alphasq))))
```

```python
kappa=1
g=10
alpha = np.linspace(0, 0.15, 100)
detrange = np.linspace(-20, 20, 200)
drives = np.array([[drive(a, det, kappa, g) for det in detrange] for a in alpha])

fig, ax = plt.subplots()
mble = ax.contour(detrange, alpha, drives, levels=np.linspace(4, 5.1, 10), cmap='viridis')
fig.colorbar(mble, label='Drive Strength')
ax.set_xlabel('$\omega_c-\omega_d$')
ax.set_ylabel('$\left|A\\right|^2$')
ax.set_title('Semiclassical Spontaneous Dressed State Polarisation', loc='right', fontdict={'fontsize': 16, 'verticalalignment': 'bottom'})
```

```python
c_q_det = (8.1831-10.5665)
kappa = 0.0024
gamma = 0
coupling_strength = 0.3347
cavity_freq = 10.5665

xiC1= (abs(c_q_det)*kappa)**(3/2)/(3**(3/4)*coupling_strength**2)
chi0=coupling_strength**2/abs(c_q_det)

def drive(A, omega_cavity, omega_drive, sigmaz, det, g, kappa):
  '''self-consistently determine xi from A'''
  def chi(A):
    return sigmaz*(g**2)/np.sqrt(2*g**2*(A**2+sigmaz)+det**2)
  return np.sqrt(A**2 * (1/omega_cavity**2)*(
            (omega_drive**2-(omega_cavity-chi(A))**2)**2 + kappa**2*omega_drive**2))
```

```python
As = np.linspace(0, 100)
Dets = np.linspace(-0.01, 0.05, 200)
Z = np.array([[drive(A, 
                  cavity_freq, 
                  cavity_freq+d, 
                  -1, 
                  c_q_det, 
                  coupling_strength, 
                  kappa) for d in Dets] for A in As])
plt.contour(Dets, As, Z, 100, linewidths=1.0)
plt.colorbar(label='Drive Strength')
plt.title('Semiclassical Dispersive Bistability', loc='right', fontdict={'fontsize': 16, 'verticalalignment': 'bottom'})
plt.xlabel('$\omega_c - \omega_d$')
plt.ylabel('$A^2$')
sb.despine()
```

### Analytic Wigner Function
http://iopscience.iop.org/article/10.1088/1464-4266/1/2/005/pdf

```python
eps = 12
x = np.linspace(-8, 8, 100)
y = np.linspace(-8, 8, 100)

def W(alpha, lamb):
    return np.exp(
        -2*np.absolute(alpha)**2) * (
        np.absolute(jv(lamb-1, np.sqrt(-8 * eps * np.conj(alpha)))/
        np.conj(alpha)**((lamb-1)/2)))**2 

w = np.array([[W(x+y*1j, 5) for x in x] for y in y])

plt.contourf(x, y, np.real(w))
plt.colorbar()
plt.title('Wigner Function', loc='right', fontdict={'fontsize': 16, 'verticalalignment': 'bottom'})
plt.xlabel('Re(Q)')
plt.ylabel('Im(Q)')
```

### Spontaneous dressed state polarisation

```python
def alpha_ss(epsilon, kappa=0.5, g=5):
    return (epsilon/kappa)* np.sqrt(1-(g/(2*epsilon))**2)

x = np.linspace(0.1, 4.5, 100)
y = np.vectorize(alpha_ss)(x)

plt.plot(x, y)
```
